```python
import torch as pt
import tensorflow as tf
```

```python
# membuat vector x ukuran 3x1 menggunakan torch dan tensorflow
# torch
x_pt = pt.tensor([[14],
                  [1] ,
                  [4]])
x_pt
```

```python
 # tensorflow
x_tf = tf.Variable([[14],
              [1] ,
              [4]])
x_tf
```

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t menggunakan torch dan tensorflow
```

```python
#torch
x_pt_t = x_pt.T
x_pt_t
```

```python
 #tensorflow
x_tf_t = tf.transpose(x_tf)
x_tf_t
```

```python
# tampilkan ukuran dari vector x_t menggunakan torch dan tensorflow
```

```python
x_pt_t.shape #torch
```

```python
x_tf_t.shape #tensorflow
```
