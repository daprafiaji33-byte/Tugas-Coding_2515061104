```python
import numpy as np
```

```python
x = np.array([[14, 1, 4],
              [4, 1, 14]])
x
```

```python
# tampilkan ukuran dari matrix X
x.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
x[:,0]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
x[1,:]
```

```python
#ulangi soal nomor 1 menggunakan pytorch
import torch as pt
```

```python
# buat matrix X berukuran 2x3 menggunakan torch
x_pt = pt.tensor([[14, 1, 4],[4, 1, 14]])
x_pt
```

```python
# tampilkan ukuran dari matrix X
x_pt.shape
```

```python
# tampilkan seluruh elemen dari kolom pertama dari matrix X
x_pt[:,0]
```

```python
# tampilkan seluruh elemen dari baris kedua dari matrix X
x_pt[1,:]
```
