Tensor Flow

```python
import tensorflow as tf
```

```python
#buat variable x
x_tf = tf.Variable(410)
```

```python
#tampilkan tipe data dari x
x_tf.dtype
```

```python
#buat variable y
y_tf = tf.Variable(14)
y_tf
```

```python
#tampilkan tipe data dari y
y_tf.dtype
```

```python
#tambahkan x+y
jumlah = tf.add(x_tf, y_tf)
jumlah
```

```python
#buat variabel bernama x_float dan simpan suatu nilai desimal
x_tf_float = tf.Variable(0.14)
x_tf_float
```

```python
#jumlahkan x_float dengan y
y_tf = tf.Variable(14, dtype=tf.float32)
jumlah_float = tf.add(x_tf_float, y_tf)
jumlah_float
```

```python
#tampilkan tipe data dari x_float
x_tf_float.dtype
```
