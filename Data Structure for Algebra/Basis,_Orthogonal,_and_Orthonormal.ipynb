```python
#buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
```

```python
import numpy as np
```

```python
i = np.array([1,0])
i
```

```python
j = np.array([0,1])
j
```

```python
#buktikan bahwa vector i dan vector j itu orthogonal
np.dot(i,j) # jika hasil dot product = 0 maka kedua vector orthogonal
```

```python
#Bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1,2])
v
```

```python
w = np.array([2,-1])
w
```

```python
np.dot(v,w)
```
